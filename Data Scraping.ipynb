```python
import requests
from bs4 import BeautifulSoup
import csv
import os

base_url = 'https://www.azlyrics.com/'

url = base_url+'j/jcole.html'  # Replace with the URL of the website you want to scrape
response = requests.get(url)

if response.status_code == 200:
    webpage_content = response.text
    soup = BeautifulSoup(webpage_content, 'html.parser')
    
    # Find all div elements with class "listalbum-item"
    items = soup.find_all('div', class_='listalbum-item')
    
    # Open a CSV file for writing
    with open('song_list.csv', mode='w', newline='',encoding='utf-8') as file:
        writer = csv.writer(file)
        
        for item in items:
            # Get text within the div element
            text = item.get_text()
            
            # Get links (anchor tags) within the div element
            links = item.find_all('a')
            
            for link in links:
                link_text = link.get_text()
                link_href = link['href']
                
                # Write the song name and its link to the CSV file
                writer.writerow([text, link_href])
else:
    print('Failed to fetch the webpage')
```
